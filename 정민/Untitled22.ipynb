```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

# Windows   
plt.rcParams['font.family'] = 'Malgun Gothic'  # ' '
plt.rcParams['axes.unicode_minus'] = False  #   

#  
questions = [
    "    ",
    "  ??",
    "  ?",
    "  ",
    "   ?"
]

# F1 Score 
f1_gpt = [0.8751, 0.8510, 0.8361, 0.8494, 0.8422]
f1_rag = [0.9851, 0.8680, 0.8618, 0.9192, 0.9227]

#   
improvement = [r - g for g, r in zip(f1_gpt, f1_rag)]

#  
df = pd.DataFrame({
    "": questions,
    " GPT": f1_gpt,
    "RAG  GPT": f1_rag,
    " ": improvement
})

# 
plt.figure(figsize=(10, 6))
bar_width = 0.35
y = range(len(df))
colors = ['#d0d0d0', '#a5d8ff']  # , 

plt.barh([i + bar_width/2 for i in y], df[" GPT"], height=bar_width, color=colors[0], label=" GPT")
plt.barh([i - bar_width/2 for i in y], df["RAG  GPT"], height=bar_width, color=colors[1], label="RAG  GPT")

#  
plt.yticks(ticks=range(len(df)), labels=df[""], fontsize=11)
plt.xlabel("BERTScore (F1)", fontsize=12)
plt.title(" GPT vs RAG GPT    (F1 )", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.xlim(0.82, 1.00)
plt.legend()
plt.tight_layout()
plt.show()
```
