```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

# Windows용 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # '맑은 고딕'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지

# 질문 리스트
questions = [
    "휴학 신청 방법 좀 알려줘",
    "군복학은 어떻게 해??",
    "수강신청 일정은 언제야?",
    "학사경고의 기준을 알려줘",
    "학생증 발급은 어떻게 하나요?"
]

# F1 Score 값
f1_gpt = [0.8751, 0.8510, 0.8361, 0.8494, 0.8422]
f1_rag = [0.9851, 0.8680, 0.8618, 0.9192, 0.9227]

# 성능 차이 계산
improvement = [r - g for g, r in zip(f1_gpt, f1_rag)]

# 데이터프레임 구성
df = pd.DataFrame({
    "질문": questions,
    "일반 GPT": f1_gpt,
    "RAG 기반 GPT": f1_rag,
    "성능 향상": improvement
})

# 시각화
plt.figure(figsize=(10, 6))
bar_width = 0.35
y = range(len(df))
colors = ['#d0d0d0', '#a5d8ff']  # 회색, 연파랑

plt.barh([i + bar_width/2 for i in y], df["일반 GPT"], height=bar_width, color=colors[0], label="일반 GPT")
plt.barh([i - bar_width/2 for i in y], df["RAG 기반 GPT"], height=bar_width, color=colors[1], label="RAG 기반 GPT")

# 그래프 설정
plt.yticks(ticks=range(len(df)), labels=df["질문"], fontsize=11)
plt.xlabel("BERTScore (F1)", fontsize=12)
plt.title("일반 GPT vs RAG GPT 응답 성능 비교 (F1 기준)", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.xlim(0.82, 1.00)
plt.legend()
plt.tight_layout()
plt.show()
```
